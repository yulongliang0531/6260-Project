```python
import numpy as np
import matplotlib.pyplot as plt
class mss:
    def __init__(self, m=1, rho=5, r=0.01, epsilon=1, p=1, L=1, T=1) -> None:
        self.m = m
        self.rho = rho
        self.r = r
        self.epsilon = epsilon
        self.p = p
        self.L = L
        self.T = T
    
    def rk4mass(self):
        array = [self.m]
        m = self.m
        r = self.r
        N = 100         # Number of Runge-Kutta steps
        h = (r)/N   # Size of Runge-Kutta steps
        def fmass(self, r):
            # dm/dr = 4 * pi * r^2 * rho
            fm = 4 * np.pi * (r ** 2) * self.rho
            return fm     
        tpoints = np.arange(0,self.r,h)
        step = 1
        for i in tpoints[1:]:
            k1 = h * fmass(self, array[step-1])
            k2 = h * fmass(self, array[step-1]+0.5*k1)
            k3 = h * fmass(self, array[step-1]+0.5*k2)
            k4 = h * fmass(self, array[step-1]+k3)
            array.append(array[step-1] + (k1+2*k2+2*k3+k4)/6)
            step += 1
        plt.plot(tpoints, array)
        plt.xlabel("radius")
        plt.ylabel("mass")
        plt.show()
        return

    def rk4p(self):
        array = [self.p]
        m = self.m
        r = self.r
        N = 100         # Number of Runge-Kutta steps
        h = (r)/N   # Size of Runge-Kutta steps
        def fpressure(self, r):
            G = 6.674e-11
            fp = -G * m * self.rho / (r ** 2)
            return fp     
        tpoints = np.arange(0,self.r,h)
        step = 1
        for i in tpoints[1:]:
            k1 = h * fpressure(self, array[step-1])
            k2 = h * fpressure(self, array[step-1]+0.5*k1)
            k3 = h * fpressure(self, array[step-1]+0.5*k2)
            k4 = h * fpressure(self, array[step-1]+k3)
            array.append(array[step-1] + (k1+2*k2+2*k3+k4)/6)
            step += 1
        plt.plot(tpoints, array)
        plt.xlabel("radius")
        plt.ylabel("pressure")
        plt.show()
        return
    
    def rk4L(self):
        array = [self.L]
        m = self.m
        r = self.r
        N = 100         # Number of Runge-Kutta steps
        h = (r)/N   # Size of Runge-Kutta steps
        def fL(self, r):
            fl = 4 * np.pi * (r**2) * self.rho * self.epsilon
            return fl     
        tpoints = np.arange(0,self.r,h)
        step = 1
        for i in tpoints[1:]:
            k1 = h * fL(self, array[step-1])
            k2 = h * fL(self, array[step-1]+0.5*k1)
            k3 = h * fL(self, array[step-1]+0.5*k2)
            k4 = h * fL(self, array[step-1]+k3)
            array.append(array[step-1] + (k1+2*k2+2*k3+k4)/6)
            step += 1
        plt.plot(tpoints, array)
        plt.xlabel("radius")
        plt.ylabel("luminosity")
        plt.show()
        return
    
    def rk4T(self):
        array = [self.T]
        m = self.m
        r = self.r
        N = 100         # Number of Runge-Kutta steps
        h = (r)/N   # Size of Runge-Kutta steps
        def fT(self, r):
            sb = 5.67e-8
            ft = - (3/16) * (self.L * self.p) / (4 * np.pi * r**4 * sb * self.T**3)
            return ft     
        tpoints = np.arange(0,self.r,h)
        step = 1
        for i in tpoints[1:]:
            k1 = h * fT(self, array[step-1])
            k2 = h * fT(self, array[step-1]+0.5*k1)
            k3 = h * fT(self, array[step-1]+0.5*k2)
            k4 = h * fT(self, array[step-1]+k3)
            array.append(array[step-1] + (k1+2*k2+2*k3+k4)/6)
            step += 1
        plt.plot(tpoints, array)
        plt.xlabel("radius")
        plt.ylabel("temperature")
        plt.show()
        return
```

```python
c = mss()
c.rk4mass()
c.rk4p()
c.rk4L()
c.rk4T()
```
